```python
import pandas as pd
import numpy as np
```

```python
def upsamling_minority_class(features, target):
    """
    Upsampling minority class to get balanced data set.

    :param target:
    :param features:
    :return:
    """
    print('Upsampling minority class')
    # indices of each class
    minor_class = np.where(target == -1)[0]
    major_class = np.where(target == 1)[0]

    n_minor_class = len(minor_class)
    n_major_class = len(major_class)
    print(f'Len of minor class is: {n_minor_class} \n and major class: {n_major_class}' )
    print("")

    # randomly choosing observation from minority class and duplicate it
    minor_class_upsampled = np.random.choice(minor_class, size=n_major_class, replace=True)

    # join together target vectors
    upsampled_target = np.concatenate((target[minor_class_upsampled], target[major_class]))

    # join together features vector
    upsampled_features = np.vstack((features[minor_class_upsampled, :], features[major_class, :]))
    return upsampled_features, upsampled_target
```
